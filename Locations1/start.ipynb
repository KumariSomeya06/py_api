```python
loc = []
my_api=[]
arr_time = []
arr_dist = []
with open('locations1.json') as file:
    js = json.load(file)
    for i in range(len(js[0]["locations"])):
        attr = js[0]['locations'][i]['attributes']
        lat = attr['lat']
        lon = attr['lon']
        loc.append([lat,lon])
num=0
flag =0
df = pd.json_normalize(js,record_path=['locations'])
while(num!=len(df)):
    api1 = f"https://matrix.route.ls.hereapi.com/routing/7.2/calculatematrix.json?start0=41.77505,-88.2742&destination{num}={loc[num][0]},{loc[num][1]}&mode=fastest%3Btruck&summaryattributes=all&apiKey=6ao6MQ1v9au2jCwQ3RK_XiwOT0f3FGlvC5RFJlhNTO0"
    response = requests.get(api1)
    daa = response.json()
    num = num+1
    if flag!=1:
        data = daa['response']['matrixEntry']
        flag =1
    for i in data:
        arr_dist.append(i['summary']['distance'])
        arr_time.append(i['summary']['travelTime'])
print(arr_dist,arr_time)
```
